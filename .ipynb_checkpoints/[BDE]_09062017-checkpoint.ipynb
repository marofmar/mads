```python
import threading
```

```python
x = 0
def inc():
    global x
    for i in range(1000000):
        x = x +1
```

```python
def dec():
    global x
    for i in range(1000000):
        x = x-1
```

```python
t1 = threading.Thread(target = inc)
t2 = threading.Thread(target = dec)
```

```python
t1.start()
```

```python
t2.start()
```

```python
t1.join()
```

```python
t2.join()
```

```python
import threading
```

```python
def worker(num):
    print("Worker: %s\n" %num)
    return
```

```python
threads = []
for i in range(3):
    t = threading.Thread(target = worker, args = (i, ))
    threads.append(t)
    t.start()
print(threading.active_count())
```

```python
import time
def say(msg):
    while True:
        time.sleep(1)
        print(msg)
        
for msg in ['you','need','python']:
    t = threading.Thread(target = say, args = (msg,))
    t.daemon = True
    t.start()
    
for i in range(100):
    time.sleep(0.1)
    print(i)
```
