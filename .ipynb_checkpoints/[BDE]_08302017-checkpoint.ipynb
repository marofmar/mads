```python
ps = Digraph(name = 'Pet Shop', node_attr = {'shape' : 'plaintext'})
```

```python
ps.node('parrot')
ps.node('alive')
ps.edge('parrot','alive')
```

```python
ps.edge_attr.update(arrowhead = 'vee', arrowsize = '2')
```

```python
from graphviz import *
```

```python
ni =Graph('ni')
```

```python
ni.attr('node', shape = 'rarrow')
ni.node('1', '@i!')
ni.node('2', 'Ni!')
ni.node('3','Ni!', shape = 'egg')
```

```python
ni.attr('node', shape = 'star')
ni.node('4','Ni!')
ni.node('5','Ni!')
ni.attr(rankdir = 'LR')
ni.edges(['12','23','34','45'])
```

```python
ni
```

```python
p = Graph(name = 'parent')
p.edge('spam','eggs')
```

```python
c = Graph(name = 'child', node_attr = {'shape': 'box'})
c.edge('foo','bar')
```

```python
p.subgraph(c)
```

```python
p
```

```python
p = Graph(name = 'parent')
p.edge('spma','eggs')
```

```python
with p.subgraph(name = 'child', node_attr = {'shape': 'box'}) as c:
    c.edge('foo','bar')
```
