```python
# change_directory.py

from os import *

home = environ['HOME']
try:
    chdir(home)
except PermissionError:
    chdir(path.split(home)[0])

dir_sel = -1
curr_path = getcwd()

def change_directory():
    global curr_path
    global dir_sel
    dir_sel = -1
    while dir_sel != 'q':
        print("change directory")
        print("------------------------------------------")
        dir_sel = input("enter a directory or path\n(enter '..' for the parent dir.(enter 'q' to exit) : ")
        switcher = {
            '..': go_to_parent_dir,
            'q': quit_change_dir
        }
        selected_func = switcher.get(dir_sel, go_to_curr_path)
        selected_func()
    return

def go_to_curr_path():
    global curr_path
    global dir_sel
    try:
        chdir(path.join(curr_path, dir_sel))
        print("current dir : %s" % path.join(curr_path, dir_sel))
        print("------------------------------------------")
        for i in listdir(path.join(curr_path, dir_sel)):
            print(i)
        print("------------------------------------------")
        curr_path = path.join(curr_path, dir_sel)
        chdir(curr_path)


    except PermissionError:
        print("------------------------------------------")
        print("permission error. go to the previous directory")
        print("------------------------------------------")
        print("current dir : %s" % curr_path)
        print("------------------------------------------")
        for i in listdir(curr_path):
            print(i)
        print("------------------------------------------")

    # except:
    #     print("cannot find the path. put an effective path.")
    #     print("------------------------------------------")

def go_to_parent_dir():
    global curr_path
    print("current dir : %s" % path.split(curr_path)[0])
    print("------------------------------------------")
    for i in listdir(path.split(curr_path)[0]):
        print(i)
    print("------------------------------------------")
    curr_path = path.split(curr_path)[0]

def quit_change_dir():
    return
```
