```python
import numpy as np
import pandas as pd
df = pd.read_csv("german_predictions.csv").set_index('OBS#')
df
```

```python
X, y = df.drop('AMOUNT',axis = 1),df.AMOUNT 
# 축을 1로 놓아야 한다! 안그럼 칼럼으로 되니까! 드롭 함수는 axis 0기본이라서. 행이 되어버리기 떄문이지
print(X)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
```

```python
X_history = X.loc[:, 'HISTORY']
X_history
```

```python
X_train_history, X_test_history, y_train, y_test = train_test_split(X_history, y, test_size = 0.2)
```

```python
lr_model.fit(X=X_train_history, y=y_train)
```

```python
from sklearn.neighbors import KNeighborsRegressor
n_neighbors = 5
knn_model = KNeighborsRegressor(n_neighbors = n_neighbors, weights = 'distance')
```

```python
knn_model.fit(X_train_history, y_train)
```

```python
dummified_X_history = pd.get_dummies(X_history)
```

```python
knn_model.fit(dummified_X_history, y)
```
