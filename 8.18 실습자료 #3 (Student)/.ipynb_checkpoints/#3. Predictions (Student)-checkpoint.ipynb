### 2. Linear Regression
---

#### 2-1. `sklearn.linear_model.LinearRegression`
---

```python
# import packages
from sklearn.linear_model import LinearRegression
```

```python
# Parameters
# fit_intercept = True # 상수를 넣을 것인지 #기본이 트루
# normalize = False # 변수들을 normalize한 상태로 모델을 얻을 것인지
```

```python
# lr_model
lr_model = LinearRegression()
```

```python
lr_model.coef_
lr_model.intercept_
```

#### 2-2. Evaluation Measure 'RMS' : `sklearn.metrics.mean_squared_error`
---

```python
# import packages
from sklearn.metrics import mean_squared_error # 절대값 쓰는 것도 있긴 함.
```

```python
# error_lr_
error_lr = mean_squared_error(y_pred = y_pred_lr, y_true = y_test)
error_lr
```

```python
# rms_lr
import math
rmse_lr = math.sqrt(error_lr)
rmse_lr
```

#### 5-2.  K-Fold Cross Validation : `sklearn.cross_validation.cross_val_score`
---

```python
# import packages
# 어떤 모델을 쓸 지 결정할 때 쓴다! KNN을 쓸지 SVM 을 쓸지 등등!
# 데이터가 작을 때 쓰기 좋다. 많으면 그냥 하면 되니까! 
from sklearn.cross_validation import cross_val_score
```

```python
# set parameters
# estimator= # fit 한 모델 # 핏 할 모델!
estimator = lr_model
# cv= # 전체 데이터를 k개의 그룹으로 나누고 총 k번의 모델 training - test를 반복한다.
cv = 5
       # 이 때 각 회차마다 test set이 되는 그룹을 계속 바꾼다.
# scorer= # Evaluation Measure로 사용할 Score Function - scorer(estimator, X, y) 무얼로 평가할거냐
scorer = rmse_scorer # 함수를 만들어준다
```

```python
# def rms_scorer(estimator, X, y)
def rmse_scorer(estimator, X, y):
    y_pred = estimator.predict(X=X) # 조금 더 추상화 시켜서 그냥 X
    rmse = math.sqrt(mean_squared_error(y_pred = y_pred, y_true = y)) # 조금 더 추상화 시켜서 적은 것
    return rmse
```

```python
# cross_val_score
cross_val_score(estimator = estimator, cv = cv, scoring = scorer, X = X, y =y) # 전체 데이터를 던져준다
```

```python
# cv_test_result_df
cv_test_result_df = pd.DataFrame()
cv_test_result_df[estimator] = cross_val_score(estimator = estimator, cv = cv, scoring = scorer, X = X, y =y)
cv_test_result_df
```
