##### Preprocessing : Dummy Variables & Scaling & Change Labels

```python
# dummy varaibles : preprocessed_df
get_dummies_cols = ['CHK_ACCT', 'HISTORY', 'SAV_ACCT', 'EMPLOYMENT', 'JOB']
dummied_df = pd.get_dummies(df, columns=get_dummies_cols)

# scaling, 수치형 데이터들에 대해서! 
scaling_cols = ['DURATION', 'INSTALL_RATE', 'AGE', 'NUM_CREDITS', 'AMOUNT']
preprocessed_df = dummied_df.copy() # 새로운 변수 만들 때에는 카피하는게 좋다. 권한다.
preprocessed_df[scaling_cols] = (preprocessed_df[scaling_cols]-preprocessed_df[scaling_cols].min()) / (preprocessed_df[scaling_cols].max()-preprocessed_df[scaling_cols].min())
# min-max scaling

# change Label
preprocessed_df.RESPONSE = df.RESPONSE-1

# split X, y 
X, y = preprocessed_df.drop('RESPONSE', axis=1), preprocessed_df.RESPONSE
```

```python
X
```

```python
y
```

##### Train-Test Split : `sklearn.model_selection.train_test_split`

```python
# import packages
from sklearn.model_selection import train_test_split

# train_test_split with 'stratify' 가지고 있는 분류의 라벨잉, 7:3 이라면, 일정한 비율 유지 위해서.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, stratify = y, random_state = 23)
```

##### `sklearn.linear_model.LogisiticRegression`

```python
from sklearn.linear_model import LogisticRegression

# parameters
class_weight = 'balanced' # 비율이 높은 애들은 샘플 적게, 비율 낮은 애들 샘플 많이 해서, 0.5로 균등하게 balanced 의도
# model
model = LogisticRegression(class_weight = class_weight) # initializing. 빈 깡통 만들기
model
# fit
model.fit(X = X_train, y = y_train)
```

```python
from sklearn.metrics import confusion_matrix

def result(fitted_model, X_test, y_test):
    # pred 
    y_pred = fitted_model.predict(X = X_test)
    # score
    score = fitted_model.score(X = X_test , y = y_test)
    # conf_mat
    conf_mat = confusion_matrix(y_true= y_test, y_pred = y_pred)
    
    return [y_pred, score, conf_mat]
```

```python
# result
raw_result = result(model, X_test, y_test)
raw_result
```

#### 1-2. RFE : Recursive Feature Elimination `sklearn.feature_selection.RFE`

```python
# import packages # like backward selection (similar)
# 계수들 중 가장 영향력 없는 애들 제거. 계수......... stepwise reg. 와는 무엇이 다른지 찾아보렴. 
# cf. RFCV
from sklearn.feature_selection import RFE
```

```python
# estimator & selector
selector = RFE(model)
```

```python
# fit to selector
selector.fit(X_train, y_train)
```

```python
# n_features_, ranking, support
print(selector.n_features_) # out of all 48 variables
print(selector.ranking_) # 선택을 했으면 1등. 1등이라서 선택한거니까. 25 제일 먼저 빠지고, 다음에 24 빠지고. 이러한 순서. 
print(selector.support_)
print(selector.estimator_)
```

```python
# change X to selected_X_train, selected_X_test
# selected_cols = 
selected_cols = selector.support_
print(selected_cols)
X_RFE_test = X_test.loc[:, selected_cols] 
X_RFE_test
```

```python
# fitting new model
RFE_model = selector.estimator_

# get result
result(RFE_model, X_RFE_test, y_test)
```

#### 1-3. PCA : `sklearn.decompostion.pca`

```python
# import packages
from sklearn.decomposition import pca
```

```python
# parameter : n_components
n_components = [5,10,15,20]
```

```python
# pca_model = 
from sklearn.linear_model import LogisticRegression
pca_model = LogisticRegression(class_weight = class_weight)

for n in n_components:
    # decomposer
    decomposer = pca.PCA(n_components=n)
    # fit
    decomposer.fit(X = X_train)
    # .transform
    X_pca_train = decomposer.transform(X_train) # 학습한 PCA 방법대로, X-train을 돌려준다. 새로운 X_train이 되는것!!!!
    #print(X_pca_train)
    print(n, sum(decomposer.explained_variance_ratio_))
    # 참고: print(n, decomposer.explained_variance_ratio_)
    
    # fit new estimator
    pca_model.fit(X = X_pca_train, y = y_train)
    # result()
    X_pca_test = decomposer.transform(X_test)
    pca_result = result(pca_model, X_pca_test, y_test)
    print(pca_result)
```
