### 1. Data
---

#### 1-1.  `german_classification.csv`
---
`german_classification.csv`는 독일 은행에 신용거래를 신청한 신청자 1000명의 데이터이다. 관찰번호(OBS#)를 제외하고 총 29개의 변수로 이루어져있으며, 각 신청자들의 '신용 등급'이 예측하고자 하는 변수이다. 신용 등급은 '우량(1)' 혹은 '불량(2)'으로 나누어지며, 각각 700명, 300명으로 구성되어 있다. 이 분석의 목적은 불량 등급일 사람을 미리 예측하여, 그들에게 신용 대출을 해주지 않도록 하는 것이다.

라벨링이 Imbalanced 되어 있다. <br>
불량은 1, 아니면 0. 관심이 있는 쪽에 1을 두는게 일반적이다. <br>


```python
import pandas as pd
import numpy as np
```

```python
# 변수별 설명 : CodeList
pd.read_csv('CodeList-classification.csv').iloc[:, :4]
```

```python
# df
df = pd.read_csv('german_classification.csv').set_index('OBS#')
```

#### 1-2. Preprocessing : Dummy Variables & Scaling & Change Labels

```python
# dummy varaibles : preprocessed_df
get_dummies_cols = ['CHK_ACCT', 'HISTORY', 'SAV_ACCT', 'EMPLOYMENT', 'JOB']
dummied_df = pd.get_dummies(df, columns=get_dummies_cols)
dummied_df
```

로지스틱 리그레션은 scaling을 해야한다. Scaling을 인덱스로 한다. 

```python
# scaling : min-max
# scaling_cols
scaling_cols = ['DURATION','INSTALL_RATE','AGE','NUM_CREDITS','AMOUNT']
preprocessed_df = dummied_df.copy()
preprocessed_df[scaling_cols] = (preprocessed_df[scaling_cols]  - preprocessed_df[scaling_cols].min())/(preprocessed_df[scaling_cols].max()-preprocessed_df[scaling_cols].min())
```

```python
preprocessed_df[scaling_cols]
```

```python
preprocessed_df.RESPONSE = df.RESPONSE-1 # 뺴기 1 , 보기 좋으라고
preprocessed_df.RESPONSE
```

```python
# change Label
```

```python
# split X, y  # axis=1 꼭 빼야해욥
X, y = preprocessed_df.drop('RESPONSE', axis = 1), preprocessed_df.RESPONSE
X
```

#### 1-3. Train-Test Split : `sklearn.model_selection.train_test_split`
---
지도학습을 위해 주어진 데이터를 train set과 test set으로 나눈다. 이 때, y의 class 비율 (class imbalance)을 고려해서 나누어준다.

```python
# import packages
from sklearn.model_selection import train_test_split
```

```python
# train_test_split with 'stratify'
# Stratify가 무엇인가! --- 클래시피케이션에서만 사용되는 개념! 
# 클래스가 다를때, 우리가 원하는 클래스의 비율을 맞춰서 뽑아준다. 
# 비율을 맞춰서 뽑아준다고! 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 77, stratify =y) # y의 값 비율에 맞춰서 트레인 테스트 분리해준다.
```

### 2. Logistic Regression
---

#### 2-1. `sklearn.linear_model.LogisticRegression`
---

```python
# import packages
from sklearn.linear_model import LogisticRegression
```

```python
# Parameters
class_weight = 'balanced'# 매우 중요한 개념.  stratify보다 모델 학습에 더 직접적으로 연관되어 있음.
'''반반 똑같이 해준다. 예를 들어 100개 중 99개가 사기, 1개가 정상이면, 99개 사기에 대해선 학습을 잘하지만 정상에 대해서는 잘 하지 못하기 떄문.'''
# fit_intercept = True # 상수를 넣을 것인지
multi_class='ovr' # 'multinomial'
''' 멀티클래스는 3개 이상의 라벨을 가지는것. 
multinomial은 해당 클래스에서 이 데이터가 0 1 2 각각을 가질 확률이 나와서, 이들 중 가장 확률이 높은 것을 선택하게 된다.
ovr 의 경우는 모델을 3개 학습해서 0/ 0이 아니다, 1/1 아니다ㅣ, 2/2 아니다, 해서 셋 중 가장 높은 확률을 선택한다. 
기본적으로 ovr로 설정되어 있음. '''
```

```python
# what does 'balanced' mean?
import numpy as np

n_samples = len(y)
n_classes = len(np.unique(y)) #클래스 갯수 0, 1 이렇게 2개
class_weight = n_samples/(n_classes*np.bincount(y)) 
'''np.bincount() 각 클래스 갯수를 세서 어레이로 반환'''
```

```python
(1000/(2*np.bincount(y)))*700 # 2는 클래스의 갯수. 1은 원래 300개였는데 500으로 갯수가 불어남. 
```

```python
# lr_model
lr_model = LogisticRegression(class_weight = class_weight)
lr_model
```

```python
# fit to model
lr_model.fit(X_train,y_train)
```

```python
# get predictions
lr_model.predict(X_test)
```

```python
# get accuracy
lr_model.score(X=X_test,y=y_test)
```

```python
# result_y_test_df 
result_y_test_df = pd.DataFrame(y_test)
result_y_test_df["lr"] = lr_model.predict(X_test)
result_y_test_df
```

#### 2-2. Evaluation : Confusion Matrix & ROC Curve
---

```python
# import packages for confusion matrix
from sklearn.metrics import confusion_matrix
```

```python
# confusion_matrix
confusion_matrix(y_pred = lr_model.predict(X_test), y_true = y_test)
# [[ true negative , false positive],
# [false negative], [true positive]]
#TN+TF/전체 = score
```

```python
def plot_roc_curve(fpr, tpr, roc_auc):
    plt.figure()
    lw = 2
    plt.plot(fpr, tpr, color='darkorange',
             lw=lw, label='ROC curve (area = {0: .2f})'.format(roc_auc))
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show()
```
