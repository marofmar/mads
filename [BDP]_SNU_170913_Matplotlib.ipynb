```python
N = 5
menMeans = (20,35,43,55,27)
womenMeans = (25,24,36,54,36)

ind = np.arange(N)
width = 0.35

p1 = plt.bar(ind-width/2, menMeans, width, color = 'Cyan')
p2 = plt.bar(ind+width/2, womenMeans, width)
```

## Side-by-side Barplot

```python
p1 = plt.bar(ind-width/2, menMeans, width, color = 'Cyan')
p2 = plt.bar(ind+width/2, womenMeans, width)
plt.ylabel('Score')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1','G2','G3','G4','G5'))
plt.yticks(np.arange(0,81,10))
plt.legend((p1[0], p2[0]), ('Men','Women')) #여기서 p1[0] 이건 무슨뜻일깜
```

```python
p1 = plt.bar(ind, menMeans, width, color = 'Cyan')
p2 = plt.bar(ind, womenMeans, width, bottom = menMeans) # 이렇게 하면 두 population의 크기를 알 수 있지
plt.ylabel('Score')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1','G2','G3','G4','G5')) #서진욱 데이터 비주얼라이제이션
plt.yticks(np.arange())
plt.legend((p1[0], p2[0]), ('Men','Women')) #여기서 p1[0] 이건 무슨뜻일깜
```

```python
spread = np.random.rand(50)*100
center = np.ones(25)*50 # 원래 50쯤 돼야할텐데, 25를 50개 만들었기에 median이 아래로 내려갔다
filter_high = np.random.rand(10)*100 + 100
filter_low = np.random.rand(10)*-100
data = np.concatenate((spread, center, filter_high, filter_low),0) #여기 9이 되면 밑으로 붙어서 3*2mat
```

```python
plt.boxplot(data)
```

```python
plt.boxplot(data, 0, 'rD') # red Diamond
```

```python
plt.boxplot(data, 0, '') # 아웃라이어 안그리기
```

```python
plt.boxplot(data, 0,'rs', 0) # vertical? horizontal four
```

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1 = np.arange(0.0,5.0, 0.1)
t2 = np.arange(0.0,5.0,0.02)
plt.figure(1)
plt.subplot(211)
plt.plot(t1,f(t1), 'bo', t2,f(t2),'k')
plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*2),'r--')
```
